QUESTION 1:xplain the difference between Relative , Absolute and Mixed Cell Referencing.


  ANSWER->
  -> RELATIVE CELL REFERENCING: A relative cell reference adjusts its position when copied to a new location. It refers to a cells position relative to the one containing the formula.

   -> ABSOLUTE CELL REFRENCING: An absolute cell reference does not change the referred cell when you drag down the Fill Handle or copy the formula.

   ->MIXED CELL REFRENCING: A mixed cell reference is a blend of fixed (absolute) and relative references.
